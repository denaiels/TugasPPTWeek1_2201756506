2. Explore the data, for instance:
    - Find minimum or maximum of a value
    - Eliminate null value
    - Replace null value to certain value

```python
# import data

import pandas as pd
listingsData = pd.read_csv("listings.csv", delimiter=",")

print(listingsData)
```

```python
# find minimum value

minValue = listingsData.min()

print(minValue)
```

```python
# find maximum value

maxValue = listingsData.max()

print(maxValue)
```

```python
# cek ada null atau tidak

print(listingsData.isna().values.any())

print(listingsData.isnull())
```

```python
# eliminate null value

newListingsDataDrop = listingsData.dropna()

print(newListingsDataDrop.isna().values.any())

print(newListingsDataDrop)
```

```python
# replace semua yang nilainya null dengan value "0"

newListingsDataFill = listingsData.fillna(0)

print(newListingsDataFill.isna().values.any())
print(newListingsDataFill)
```
